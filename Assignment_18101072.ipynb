```
class Musical_instrument:
  def __init__(self,iname):
    self.iname=iname
    print('Music instrument detected!')

class String(Musical_instrument):
  def display(self):
    print('String type music instrument')

class Wind(Musical_instrument):
  def display(self):
    print('Wind type music instrument')
  

class Violin(String):
  def function(self):
    print('Need string to use')

class Flute(Wind):
  def function(self):
    print('Need air to use')

class Piano(String,Wind):
  def function(self):
    print('Need string and air both to use')


```

```

Baroque_Violin=Violin('violin')
Baroque_Violin.display()
Baroque_Violin.function()
print('\n')

Piccolo_flute=Flute('flute')
Piccolo_flute.display()
Piccolo_flute.function()
print('\n')

Grand_Piano=Piano('piano')
Grand_Piano.display()
Grand_Piano.function()

```
